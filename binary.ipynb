```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

data= pd.read_csv('2.02. Binary predictors.csv')
data
```

Tenemos 168 datos de personas que admitidas a la universidad según sus puntajes de SAT y su género, revisaremos las posibilidades de admisión según las variables presentadas y el modelo de regresión logistica.


Al usar más variables podemos obtener mayor precisión, podremos ver a continuacion las diferencias, primero usaremos solo la variable género y despues usaremos género y SAT...

```python
df=data.copy()
df['Admitted']=df['Admitted'].map({'Yes':1, 'No':0})
df['Gender']=df['Gender'].map({'Female':1, 'Male':0})
df
```

```python
y= df['Admitted']
X= df['Gender']
```

```python
x= sm.add_constant(X)
regresion=sm.Logit(y,x).fit()
```

```python
regresion.summary()
```

```python
np.exp(2.0786)
```

La posibilidad de admisión de una mujer es 7.99 veces mayor que un hombre, en este caso solo usamos la variable Genero...

A continuación emplearemos el Genero y el SAT...

```python
df1=data.copy()
df1['Admitted']=df1['Admitted'].map({'Yes':1, 'No':0})
df1['Gender']=df1['Gender'].map({'Female':1, 'Male':0})
df1
```

```python
dep=df1['Admitted']
indep=df1[['Gender','SAT']]
```

```python
ind=sm.add_constant(indep)
modelo=sm.Logit(dep,ind).fit()
```

```python
modelo.summary()
```

Esta tabla de resultados nos permite ver que ahora la Log verosimilitud o Log likelihood es mucho mayor que cuando solo usamos la variable de Género para predecir la admisión a la universidad. Anteriomente fue el valor arrojado fue de  -96.140 y ahora es de -20.180, esto significa que la variable SAT es significativa para predecir los resultados.

```python
np.exp(1.9449)
```

Lo que vemos ahora es que la posibilidad de admisión de una mujer es 7 veces mayor que un hombre en caso de que tengan el mismo puntaje del SAT.

En este caso vemos que la universidad permite a las mujeres, mayores posibilidades de entrar a la universidad, sin embargo, podriamos encontrarno en nuestro trabajo más variables que nos indiquen más criterios de admisión en la universidad.

```python
def precision (entradas, valoresReales, modelos):

    valoresPredecidos= modelos.predict(entradas) 
    arreglo=np.array([0,0.5,1])
    cm=np.histogram2d(valoresReales, valoresPredecidos, bins=arreglo)[0]
    matriz=(cm[0,0]+cm[1,1])/cm.sum()

    return cm, matriz
```

```python
matrizConfusion=precision(prueba, test_1, modelo)
matrizConfusion
```

Al hacer la prueba con datos que el modelo no habia contemplado lo que tenemos es que:

De 12 datos donde la prediccion fue 1, los datos reales fueron 1

De 1 dato donde la prediccion fue 1, los datos reales fueron 0

Entonces hubo 17 aciertos y 2 errores, por lo tanto, la precisión ahora es de 89%. Además esto significa que disminuyo respecto a los valores que obtuvimos anteriormente, pues esta prueba implicó un sobreajuste al modelo.
